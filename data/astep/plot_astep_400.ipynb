```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
from astropy.table import Table, vstack
%matplotlib notebook
t17 = ascii.read('betapic_astep_2017.csv')
t18 = ascii.read('betapic_astep_2018.csv')


```

```python
tb = vstack([t17,t18])
```

```python
plt.scatter(tb['BJD'], tb['FCAL1'])
plt.scatter(tb['BJD'], (tb['SUNELEV']<-15))
plt.scatter(tb['BJD'], tb['FLAG']*0.95)
plt.plot(tb['BJD'], tb['SKY']/200.)
```

```python

m_elev = (tb['SUNELEV']<-15.)
m_sky  = ( (tb['SKY']/200.) < 1)

m_all = m_elev * m_sky * tb['FLAG']



m_all = (m_elev * m_sky * tb['FLAG']).astype('bool')
```

```python
plt.scatter(tb['BJD'], tb['FCAL1'])
plt.scatter(tb['BJD'], (tb['SUNELEV']<-15))
plt.scatter(tb['BJD'], m_all)
```

```python
a=tb['BJD']
b=tb['FCAL1']

plt.scatter(a[m_all], b[m_all]   )
```
